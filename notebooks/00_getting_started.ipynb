# Getting started (ArcGIS Pro Python + VS Code)

This notebook is a short “day 1” check to confirm your environment and kernel are set correctly.

Goals:
- confirm you are running Python from the `.conda-arcgis` environment
- confirm `arcpy` imports successfully

---

## Before you start

1. Create the per-repo environment (one-time) in **ArcGIS Pro Python Command Prompt**:

    conda create --name .conda-arcgis --clone arcgispro-py3

2. In VS Code, open a terminal and activate:

    conda activate .conda-arcgis

---

## Select the kernel (VS Code)

In VS Code:
- Open the Command Palette (Ctrl+Shift+P)
- Run: **Python: Select Interpreter** and choose `.conda-arcgis`
- In this notebook, click **Select Kernel** and choose the `.conda-arcgis` kernel/interpreter

---
